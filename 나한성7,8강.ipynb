Monte Carlo 시뮬레이션

4:파이=전체 점의 수:원 내부 점의 수
파이 = 4*원 내부 점의 수/전체 점의 수

```python
import random #random 모듈을 불러옴
NUMBER_OF_TRIALS=5000000
nuberOFHITs=0

for i in range(NUMBER_OF_TRIALS):
  x=random.random()*2-1 #random 모듈 안에 random() 함수가 있음, 0~1사이의 값이 랜덤하게 배정됨
  y=random.random()*2-1 #점을 랜덤하게 찍음
  if x*x+y*y<=1:
    nuberOFHITs+=1
pi=4*nuberOFHITs/NUMBER_OF_TRIALS
print("PI는",pi,"입니다.")
```

```python
fa_age=39
def up_fa_age():
  global fa_age
  fa_age+=1

def get_fa_age():
  return fa_age

print("2019년")
print("아빠:",get_fa_age())
print("2020년")
up_fa_age()
print("아빠:",get_fa_age())
```

```python
class Ageinfo:
  def up_age(self):
    self.age+=1
  def get_age(self):
    return self.age
```

```python
fa = Ageinfo() # fa 객체 생성
fa.age=39 # fa 객체 안에 age 라는 변수 생성

print("아빠의 나이:",fa.get_age())

print("1년 뒤...")
fa.up_age()
print("아빠의 나이:",fa.get_age())
```

```python
fa = Ageinfo()
mo = Ageinfo()
me = Ageinfo()

fa.age=50
mo.age=49
me.age=30

print("아빠의 나이:",fa.get_age())
print("엄마의 나이:",mo.get_age())
print("누나의 나이:",me.get_age())

print("1년 뒤...")
fa.up_age()
mo.up_age()
me.up_age()

print("아빠의 나이:",fa.get_age())
print("엄마의 나이:",mo.get_age())
print("누나의 나이:",me.get_age())
```

```python
x=np.random.rand(100)
eps=np.random.rand(100)/1.5
y=3*x+1+eps
```

```python
plt.plot(x,y,'.')
plt.show()
```

```python
a=0
b=0
alpha=0.01
n=100
err_evolve=[]
it_index=[]
for it in range(2000):
  
   delta_a=0
   delta_b=0
   err=0
   for i in range(n):
      delta_a=delta_a+2*(a*x[i]+b-y[i])*x[i]
      delta_b=delta_b+2*(a*x[i]+b-y[i])
      err=err+(a*x[i]+b-y[i])*(a*x[i]+b-y[i])/n
   a=a-alpha*delta_a/n
   b=b-alpha*delta_b/n
   
   err_evolve.append(err)
   it_index.append(it)
   
   if it%200==0:
     plt.plot(x,y,'.',x,a*x+b)
     plt.legend(labels=('Sample Data''Linear Regression'))
     plt.title('iteration %d' %it)
     plt.show()
```

```python
b
```

```python
plt.plot(it_index,err_evolve)
plt.title('Change of MSE')
plt.show()
```
